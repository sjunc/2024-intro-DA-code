산점도

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2025)    # 시드로 랜덤이지만 같은 값을 뽑을 수 있음.

N = 80
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (20*np.random.rand(N)) **2 # 0 to 15 point radii

plt.scatter(x, y, s = area, c = colors, alpha = 0.7)       #alpla 불투명도
plt.colorbar()                                              # 옆에 컬러바
plt.show()
```

```python
# 논리 인덱싱 산점도
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000)  #-100부터 100까지 1000개 랜덤 값 출력
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50    # 저장된 값 중 절대값이 50보다 큰 값 걸러냄
mask2 = abs(y) > 50
x = x[mask1 + mask2]    # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2]

plt.scatter(x, y, c =x, cmap = 'jet', alpha = 0.4)
plt.colorbar()
plt.show()
```

### 돗수분포표 시험문제

rwidth 사이간격

bins 인자 혹은 구간

bins = [.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5 ]      = bins = np.arange(.5,7)

0.5~1.5 까지 첫 구간 1.5 ~ 2.5  

bins 기본 값은 최소에서 최대까지 10개로 나눔. 

random.choice p로 각각의 확률을 지정 가능 11, 12, 15, 16p

```python
plt.figure(figsize = (6, 4))    #그림 크기 지정

plt.hist([1, 1, 1.1, 2, 2, 3, 4, 5, 6, 6, 6, 6], rwidth = .98, bins = [.5, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5])
plt.xticks(range(1, 7))
plt.show()
```

```python
dice = np.random.choice(range(1,7), 10000000, p = [0.15, 0.25, 0.3, 0.1, 0.1, 0.1]) # p로 확률 지정함. 모두의 합은 1이 되어야함
plt.figure(figsize = (5,3), dpi=200)
plt.hist(dice, rwidth =.98, bins = np.arange(.5, 7))
plt.xticks(range(1, 7))
plt.savefig('dice.png')
plt.show()
```
