'ro' color mark 빨간 원 'or' 순서 상관 없음. 약자가능(빨초검....)
linewidth -> lw 약어 기본 1
linestyle 지정 안하면 선이 안 그어짐. ' r .' 빨간 작은 점만
'-.m*' 마젠타색 *모양 점의 대시닷선



```python
# lecture 10 p.20
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40], 'r .' , label='circle')      # 선이 나오지 않음
plt.plot([40, 30, 20, 10], 'g^:' , label='triangle up') #:을 넣어서 점선을 그림

plt.title('marker')
plt.xlabel('x axis')
plt.xlabel('y axis')
plt.legend()
plt.show()
```
